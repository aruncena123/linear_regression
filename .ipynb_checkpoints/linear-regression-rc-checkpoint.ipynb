## Gradient Descent

## Gradient Descent finds the local minima of a function

### Equation 


"Gradient descent is a first-order iterative optimization algorithm for finding the minimum of a function. To find a local minimum of a function using gradient descent, one takes steps proportional to the negative of the gradient (or of the approximate gradient) of the function at the current point."

Here we take the derivative of two functions and check for which we can the least error.


we need to try and minimize the error as much as possible


```python
from numpy import *                                            # import all functions from numpy 
from matplotlib import pyplot as plt                           # import matplotlib to display graph
```

```python
# calculate the squared sum of errors 
def sqrd_error_for_points(b,m,points):
    total_error = 0                                            # setting total error as 0 and summing it all together later
    for i in range(0,len(points)):                             # iterating throgh out the data set
        x = points[i,0]                                        # assigning x
        y = points[i,1]                                        # assigning y
        total_error += (y - (m * x) + b) ** 2                  # applying formula
    return total_error / float(len(points))                    # returing the squared sum of errors
```

```python
# determine the gradient descent by using partial derivation
def step_gradient_descent(current_m,current_b,points,learning_rate):
    gradient_m = 0
    gradient_b = 0
    N = float(len(points))
    
    for i in range(0,len(points)):
        x = points[i,0]
        y = points[i,1]
        gradient_m += -(2 / N) * x * (y - ((current_m * x) + current_b ))
        gradient_b += -(2 / N) * (y - ((current_m * x) + current_b ))
    
    new_m = current_m - (learning_rate * gradient_m)
    new_b = current_b - (learning_rate * gradient_b)
    return [new_b,new_m]
```

```python
# determine the gradient descent of the equation
def gradient_descent_runner(points,learning_rate,number_iterations,inital_b,inital_m):
    b = inital_b
    m = inital_m
    
    for i in range(number_iterations):
        b,m = step_gradient_descent(m,b,array(points),learning_rate)
    return [b,m]
```

```python
points = genfromtxt('data.csv',delimiter=',')                  # importing all the data points from the csv file, the delimiter parces the values 
#print(points)                                                  # view points
# if possible try to view the data set as it is more intutive that way
for i in range(len(points)):                                   # iterates through the entire data set
    x = points[i,0]                                            # select all the rows in column 0
    y = points[i,1]                                            # select all the rows in column 1
    plt.scatter(x,y)                                           # scatter all points of x an d y 
plt.show()                                                     # display graph
```

```python
initial_b = 0
initial_m = 0 
# here let us assume that initially the line is horizontal and get the squared error function
# Goal is to reduce the value eg.('5565.10783448') to as minimun as possible
print('the inital error of the line is :',sqrd_error_for_points(initial_b,initial_m,points))

number_iterations = 1000                                      # number of iterations 
learning_rate = 0.0001                                         # hyper parameter(something that needs to chosen by trial and error)
[b,m] = gradient_descent_runner(points,learning_rate,number_iterations,initial_b,initial_m)
print('the new values of b and m are = ',b,m)
print('y = mx + b',"===>",'y = {0}x + {1}'.format(m,b))
print("the final error of the line is : ",sqrd_error_for_points(b,m,points))
```
